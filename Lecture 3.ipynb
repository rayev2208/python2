### Multidimensional array

Index slicing works exactly the same way for multidimensional arrays:

```
A = np.array([[n+m*10 for n in range(5)] for m in range(5)])
print(A)
```

```
print(A[1:4, 1:4]) # a block from the original array 
```

```
print(A[::2, ::2]) # strides 
```

#### Fancy indexing



Fancy indexing is the name for when an array or list is used in place of an index:

```
# Row
row_indices = [1, 2, 3]
print(A) #original array
print()
print(A[row_indices]) #changed
```

```
# Column
col_indices = [1, 2, -1] # remember, index -1 means the last element
print(A) #original array
print()
print(A[col_indices]) #changed
```

We can also use index masks: If the index mask is an Numpy array of data type bool, then an element
is selected (True) or not (False) depending on the value of the index mask at the position of each element

```
B = np.array([n for n in range(5)])
row_mask = np.array([True, False, True, False, False]) #bool array
print(B) #original array
print()
print(B[row_mask]) #changed
```

```
B = np.array([n for n in range(5)])
row_mask = np.array([1,0,1,0,0], dtype=bool) #int array converted to bool type
print(B) #original array
print()
print(B[row_mask]) #changed
```

This feature is very useful to conditionally select elements from an array, for example, by using comparison
operators:

```
x = np.arange(0, 10, 0.5)
mask = (5 < x) * (x < 7.5)
print(x)
print()
print(x[mask])
```

#### Functions for extracting data from arrays and creating arrays

The index mask can be converted to position index using the where function:

```
indices = np.where(mask)
print("Mask:", mask)
print()
print(indices)
print()
print(x[mask])
```

With the diag function we can also extract the diagonal and subdiagonals of an array:

## Numpy - Reshaping, Resizing and Stacking

The shape of an Numpy array can be modified without copying the underlaying data, which makes it a fast operation even for large arrays.

```
A
```

```
n, m = A.shape
print(n,m,'\n')

B = A.reshape((1,n*m))
print(B)
```

#### Modifying

```
B[0,0:5] = 5 # modify the array
print(B)
print()
print(A) # and the original variable is also changed. B is only a different view of the same data
```

We can also use the function flatten to make a higher-dimensional array into a vector. But this function
create a copy of the data.


```
B = A.flatten()
print(B, '\n')

B[0:5] = 10
print(B, '\n')
print(A) # now A has not changed, because Bs data is a copy of As, not refering to the same data
```

## Iterating over array elements

```
v = np.array([1,2,3,4]) 

for element in v:
  print(element)
```

```
M = np.array([[1,2], [3,4]])
for row in M: 
  print("row", row)
  for element in row:
    print(element)
```

#### enumerate:

When we need to iterate over each element of an array and modify its elements, it is convenient to use the **enumerate** function to obtain both the element and its index in the for loop:

```
for row_idx, row in enumerate(M):
  print("row_idx", row_idx, "row", row)
  for col_idx, element in enumerate(row):
    print("col_idx", col_idx, "element", element) 
    
    # update the matrix M: square each element
    M[row_idx, col_idx] = element ** 2
```

#### ndenumerate:

```
row = np.array([1, 2, 3])

for idx, x in np.ndenumerate(row):
  print(idx, x)
```

#### nditer:

The function nditer() is a helping function that can be used from very basic to very advanced iterations. It solves some basic issues which we face in iteration, lets go through it with examples.

```
row = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

for x in np.nditer(row):
  print(x)
```

We can use **op_dtypes** argument and pass it the expected datatype to change the datatype of elements while iterating.

NumPy does not change the data type of the element in-place (where the element is in array) so it needs some other space to perform this action, that extra space is called buffer, and in order to enable it in **nditer()** we pass **flags=['buffered']**.

```
row = np.array([1, 2, 3])

for x in np.nditer(row, flags=['buffered'], op_dtypes=['S']):
  print(x)
```

```
array = numpy.zeros([100, 200, 3], dtype=numpy.uint8)
array[:,:100] = [255, 128, 0] #Orange left side
array[:,100:] = [0, 0, 255]   #Blue right side

img = PIL.Image.fromarray(array)
img.save('simple.png')
```

```
array = numpy.zeros([100, 200], dtype=numpy.uint8)

# Set grey value to black or white depending on x position
for x in range(200):
    for y in range(100):
        if (x % 16) // 8 == (y % 16) // 8:
            array[y, x] = 0
        else:
            array[y, x] = 255

img = PIL.Image.fromarray(array)
img.save('testgrey.png')
```
