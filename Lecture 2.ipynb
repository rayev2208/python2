```
class Data:
  def __init__(self, filename):
    self.filename = filename

  def read_data(self, path):
    data = []
    with open(path + self.filename, 'r') as rfile:
      for row in rfile.readlines():
        data.append(row)
    return data
```

```
filenames = ['wishlist.txt', 'peoplelist.txt']
path = r"/content/drive/My Drive/Python2_new/week2/"

wish_data = Data(filenames[0]).read_data(path)
name_data = Data(filenames[1]).read_data(path)
```

```
len(wish_data), len(name_data)
```

```
import random

class RandomSender:
  def __init__(self, wish_data, name_data):
    self.wish_data = wish_data
    self.name_data = name_data

    self.name = random.choice(self.name_data)
    self.wish = random.choice(self.wish_data)

  def checker(self, folder_name, path):
    if os.path.isfile(path+folder_name+'{}.txt'.format(self.name.strip())):
      print("Name: " + self.name.strip())
      print("You have already sent wish to this guy!")
    else:
      print("Name: " + self.name.strip())
      print("You can send to this guy")

  def send_text(self, folder_name, path):
    with open(path+folder_name+'{}.txt'.format(self.name.strip()), 'w+') as wfile:
      wfile.write(self.wish)
```

```
folder_name = 'sent/'
var = True
while var:
  answer = input("Do you wanna send a wish? y/n: ")
  if answer == 'y':
    os.chdir(path)
    try:
      os.mkdir('sent')
    except:
      print("Folder is already created")
      os.chdir(path + folder_name)
      #print(os.getcwd())

      sender = RandomSender(wish_data, name_data)
      sender.checker(folder_name, path)

      choice = input("Send or not? y/n: ")
      if choice == 'y':
        sender.send_text(folder_name, path)
        print("Done.......!")
      else:
        print("Sender class is closed")

  else:
    var = False
```
